```python
import pandas as pd 
```

```python
BMresults = pd.read_csv('BMresults.csv', sep= ',')
BMresults.head()
```

```python
data = pd.read_csv('data.csv', sep = ',')
data.head()
```

```python
diffexpvalues = pd.read_csv('diffexpvalues.csv', sep = ',')
diffexpvalues.shape
```

```python
matches = pd.DataFrame()
for line in data: 
    for row in BMresults:
        if line.loc('Unnamed: 0') == row.loc('Gene stable ID version'):
            matches.append(row)
print(matches)
```

```python
BMresults.index
```

```python
data_matches = BMresults.merge(data, how = 'inner', left_on= 'Gene stable ID version', right_on= 'Unnamed: 0')
data_matches.shape
```

```python
diff_matches = BMresults.merge(diffexpvalues, how = 'inner', left_on= 'Gene stable ID version', right_on= 'Unnamed: 0')
diff_matches.head()
```

```python
unique = diff_matches['Gene stable ID version'].unique()
diff_unique = pd.DataFrame(columns=diff_matches.columns)
for line in unique:
    diff_unique = diff_unique.append(diff_matches[diff_matches['Gene stable ID version'] == line].iloc[0:1,:])
diff_unique    
```

```python
unique_data = data_matches['Gene stable ID version'].unique()
data_unique = pd.DataFrame(columns=data_matches.columns)
for line in unique_data:
    data_unique = data_unique.append(data_matches[data_matches['Gene stable ID version'] == line].iloc[0:1,:])
```

```python
data_unique.to_csv('data_unique.csv')
diff_unique.to_csv('diff_unique.csv')
```
