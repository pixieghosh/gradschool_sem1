```python
with sqlite3.connect('/Users/pixghosh/Desktop/databases/2.2.22 practice/prac.db') as con:
    race_df = pd.read_sql('select * from race', con)
    var = ""
    for row in race_df.iterrows():
        var += f"update race set race_annotation = '{row[1]['race_annotation'][::-1]}' where code = {row[1]['code']};\n"
        print(var)
    con.executescript(var)
    race_df = pd.read_sql('select * from race', con)
race_df.head()
```

```python
with sqlite3.connect('/Users/pixghosh/Desktop/databases/2.2.22 practice/prac.db') as con:
    demographics_df = pd.read_sql('select * from demographics where maritalstatus = 5', con)
demographics_df.head()
```

```python
create = '''CREATE TABLE marital_probs (
    patientid      TEXT    PRIMARY KEY,
    dateofbirth    DATE,
    maritalstatus  INTEGER REFERENCES marital (code),
    race           INTEGER REFERENCES race (code),
    ageatdiagnosis INTEGER,
    alivestatus    INTEGER REFERENCES alive (code),
    survivaltime   INTEGER
);'''

with sqlite3.connect('/Users/pixghosh/Desktop/databases/2.2.22 practice/prac.db') as con:
    con.execute(create)
demographics_df.head()

```

```python
demographics_df.to_sql("marital_probs",con,if_exists='append', index = False, index_label = "patientid")
```
